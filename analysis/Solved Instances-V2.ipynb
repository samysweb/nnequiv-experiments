```python
import parse
import analyse
import matplotlib.pyplot as plt
import numpy as np
```

```python
benchmarks = [
    "../results/ACASXU_run2a_1_1_batch_2000-trunc-1010-0.05",
    "../results/ACASXU_run2a_1_2_batch_2000-trunc-1010-0.05",
    "../results/HAR-trunc-1000.1-0.25",
    "../results/HAR-trunc-1000-0.25",
    "../results/mnist_relu_3_100-trunc-400.1-0.25",
    "../results/mnist_relu_3_100-trunc-400-0.25",
    "../results/mnist8x8_student_12_12_12_10-9000.5-top",
    "../results/mnist8x8_student_12_12_12_10-9100.5-top",
    "../results/mnist8x8_student_24_12_10-9000.5-top",
    "../results/mnist8x8_student_24_12_10-9100.5-top",
    "../results/mnist8x8_student_36_10-9000.5-top",
    "../results/mnist8x8_student_36_10-9100.5-top",
    "../results/mnist8x8_student_12_12_12_10-9001-top",
    #"../results/mnist8x8_student_12_12_12_10-9101-top"
]
variants = [
    'DONT',
    'CEGAR',
    'REFINE_UNTIL_LAST',
    'REFINE_UNTIL_MAX',
    'REFINE_UNTIL_95P'
]
commit = 'd57cd8b'
out_handlers=[]
err_handlers=[parse.RunLim]
```

```python
runs = []
for v in variants:
    cur_run = {}
    for b in benchmarks:
        try:
            cur_run[b]=parse.BenchmarkRun(
            b+"-"+v+"/"+"1/nnequiv-"+commit+"/stdout.log",
            b+"-"+v+"/"+"1/nnequiv-"+commit+"/stderr.log",
            out_handlers=out_handlers,
            err_handlers=err_handlers
            )
        except FileNotFoundError:
            print(f"Could not find {b}-{v}")
            cur_run[b]=None
    runs.append(cur_run)
```

```python
variants = [
    'REFINE_UNTIL_LAST_OPTIMISTIC1',
    'REFINE_UNTIL_LAST_OPTIMISTIC2'
]
commit = '2afb4f8'
```

```python
for v in variants:
    cur_run = {}
    for b in benchmarks:
        try:
            cur_run[b]=parse.BenchmarkRun(
            b+"-"+v+"/"+"1/nnequiv-"+commit+"/stdout.log",
            b+"-"+v+"/"+"1/nnequiv-"+commit+"/stderr.log",
            out_handlers=out_handlers,
            err_handlers=err_handlers
            )
        except FileNotFoundError:
            print(f"Could not find {b}-{v}")
            cur_run[b]=None
    runs.append(cur_run)
```

```python
buening_speedup = {}
buening_master = {}
for b in benchmarks:
    try:
        buening_speedup[b]=parse.BenchmarkRun(
            b+"/1/buening-0badb2c/stdout.log",
            b+"/1/buening-0badb2c/stderr.log",
            out_handlers=[],
            err_handlers=[parse.RunLim]
            )
    except FileNotFoundError:
        print(f"Could not find {b}")
        buening_speedup[b]=None
    try:
        buening_master[b]=parse.BenchmarkRun(
            b+"/1/buening-a5c91be/stdout.log",
            b+"/1/buening-a5c91be/stderr.log",
            out_handlers=[parse.EquivLine],
            err_handlers=[parse.RunLim]
            )
    except FileNotFoundError:
        print(f"Could not find {b}")
        buening_master[b]=None
runs.append(buening_speedup)
runs.append(buening_master)
```

```python
comparison = analyse.RunlimMultiComparator(runs,["exact", "cegar", "last", "max", "95p", "last_opt1", "last_opt2", "buening-speed", "buening-master"],benchmarks)
    
comparison.render_table("real")
```
