```python
def get_split_plot(split_hash):
    net1 = {}
    net2 = {}
    for s in split_hash.keys():
        if s[0]==0:
            if s[1] in net1:
                net1[s[1]]+=1
            else:
                net1[s[1]]=1
        elif s[0]==1:
            if s[1] in net2:
                net2[s[1]]+=1
            else:
                net2[s[1]]=1
    return net1, net2
cegar_net1, cegar_net2 = get_split_plot(overapprox_hash)
exact_net1, exact_net2 = get_split_plot(acas_exact.split.split_positions)
```

```python
def get_split_share(splits,net):
    rv = []
    for layer in net.keys():
        rv.append(splits[layer]/net[layer])
    return rv
cegar_net1_share = get_split_share(cegar_net1, net1_arch)
cegar_net2_share = get_split_share(cegar_net2, net2_arch)
exact_net1_share = get_split_share(exact_net1, net1_arch)
exact_net2_share = get_split_share(exact_net2, net2_arch)
```
